```python
import re

def remove_unicode(text):

    try:
        text = text.encode("latin_1").decode("raw_unicode_escape").encode('utf-16', 'surrogatepass').decode('utf-16')
    except:
        text = text
        
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
        "]+", re.UNICODE)
    
    text = re.sub(emoji_pattern, '', text)
    text = ''.join(text.split('\\n')) # to remove uncompiled \n
    text = re.sub('\n', '', text)     # to remove compiled \n

    return text
```

```python
# get and adjust the stop words

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('english')
# tokens_without_sw=[t for t in list_data2 if t not in stop_words]
include = ["until","against","most","before","after","more","over","under","no","same","when","why","how","any"]
[stop_words.remove(i) for i in include]
print(stop_words)
```

```python
def clean_data(text, small_word_lmt=0,stop_words = stop_words,
               extra_words=["bank","zar","us","jp","ive","got","get","amp","citi","ubs","im","zs","rit","uk"]):
    
    text = remove_unicode(text)
    list_data2=[]
    sentence=text.lower().split(' ')
    for word in sentence:
        if not (word.startswith('@') | word.startswith('#') | word.startswith('http')) \
           and len(word) > small_word_lmt \
           and word not in extra_words \
           and word not in stop_words \
           and word.isalpha():
            list_data2.append(word)

    return " ".join(list_data2)
```

```python
df.tweet = df.tweet.apply(clean_data)
```

```python
#adjustment for non-english words

nltk.download('words')
eng_words = set(nltk.corpus.words.words())
df.tweet = df.tweet.apply(lambda x: " ".join([i for i in nltk.wordpunct_tokenize(x) if i in eng_words]))
```

```python
df_clean = df.tweet
df_clean.shape
```

```python
df_clean = df_clean.dropna()
df_clean.shape
```

```python
list_data = df_clean.to_list()
list_data = [str(sent) for sent in list_data]
```

```python
from sentence_transformers import SentenceTransformer
```

```python
# distilbert-base-nli-mean-tokens
# paraphrase-distilroberta-base-v1
```

```python
model=SentenceTransformer('distilbert-base-nli-mean-tokens')
embeddings1 = model.encode(list_data, show_progress_bar=True)
```

```python
model=SentenceTransformer('paraphrase-distilroberta-base-v1')
embeddings2 = model.encode(list_data, show_progress_bar=True)
```

```python
embeddings2 = embeddings2 /  np.linalg.norm(embeddings2, axis=1, keepdims=True)
```

```python
import umap
```

```python
from sklearn.decomposition import PCA 
from sklearn.manifold import TSNE
```

```python
dim_embedding2_1 = umap.UMAP(n_components=2).fit_transform(embeddings1)
dim_embedding2_2 = umap.UMAP(n_components=2).fit_transform(embeddings2)
dim_embedding3_1 = umap.UMAP(n_components=3).fit_transform(embeddings1)
dim_embedding3_2 = umap.UMAP(n_components=3).fit_transform(embeddings2)
dim_embedding4_1 = umap.UMAP(n_components=4).fit_transform(embeddings1)
dim_embedding4_2 = umap.UMAP(n_components=4).fit_transform(embeddings2)
dim_embedding5_1 = umap.UMAP(n_components=5).fit_transform(embeddings1)
dim_embedding5_2 = umap.UMAP(n_components=5).fit_transform(embeddings2)
#n_neighbors=10, min_dist=0.0, 
```

```python
dim_PCA = PCA(n_components=2).fit_transform(embeddings2)
dim_tSNE = TSNE(n_components=2).fit_transform(embeddings2)
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import hdbscan
```

```python
def cluster_results(dim_embeddings = [dim_embedding2_1,dim_embedding2_2,dim_embedding3_1,dim_embedding3_2,
                                      dim_embedding4_1,dim_embedding4_2,dim_embedding5_1,dim_embedding5_2],
                    Kmeans_clusters=10,
                    hdbscan_min_samples = 20, hdbscan_min_cluster_size = 20,
                    Agg_num_clusters=None,Agg_distance_threshold=1.5):
    
    ClusterDict = {} 
    #these are dont have reduced dims which is 768
    ClusterDict["ClusterKmeans_1A_bert1"] = KMeans(n_clusters=Kmeans_clusters).fit_predict(embeddings1)
    ClusterDict["ClusterKmeans_1B_bert2"] = KMeans(n_clusters=Kmeans_clusters).fit_predict(embeddings2)
    ClusterDict["ClusterKmeans_PCA"] = KMeans(n_clusters=Kmeans_clusters).fit_predict(dim_PCA)
    ClusterDict["ClusterKmeans_tSNE"] = KMeans(n_clusters=Kmeans_clusters).fit_predict(dim_tSNE)
    
    ClusterDict["hdbscan_1A_bert1"] = hdbscan.HDBSCAN(min_samples = hdbscan_min_samples,
                                              min_cluster_size=hdbscan_min_cluster_size,
                                              cluster_selection_epsilon = 0.5,
                                              metric = 'euclidean',
                                              cluster_selection_method = "eom").fit_predict(embeddings1)
    ClusterDict["hdbscan_1B_bert2"] = hdbscan.HDBSCAN(min_samples = hdbscan_min_samples,
                                             min_cluster_size=hdbscan_min_cluster_size,
                                             cluster_selection_epsilon = 0.5,
                                             metric = 'euclidean',
                                             cluster_selection_method = "eom").fit_predict(dim_PCA)
    ClusterDict["hdbscan_PCA"] = hdbscan.HDBSCAN(min_samples = hdbscan_min_samples,
                                         min_cluster_size=hdbscan_min_cluster_size,
                                         cluster_selection_epsilon = 0.5,
                                         metric = 'euclidean',
                                         cluster_selection_method = "eom").fit_predict(dim_tSNE)
    ClusterDict["hdbscan_tSNE"] = hdbscan.HDBSCAN(min_samples = hdbscan_min_samples,
                                         min_cluster_size=hdbscan_min_cluster_size,
                                         cluster_selection_epsilon = 0.5,
                                         metric = 'euclidean',
                                         cluster_selection_method = "eom").fit_predict(embeddings2)
    
    ClusterDict["ClusterAgglomerative_1A_bert1"] = AgglomerativeClustering(n_clusters=Agg_num_clusters, 
                                                    distance_threshold=Agg_distance_threshold,
                                                    affinity='cosine',
                                                    linkage='average').fit_predict(embeddings1)
    ClusterDict["ClusterAgglomerative_1B_bert2"] = AgglomerativeClustering(n_clusters=Agg_num_clusters, 
                                                distance_threshold=Agg_distance_threshold,
                                                affinity='cosine',
                                                linkage='average').fit_predict(embeddings2)
    ClusterDict["ClusterAgglomerative_PCA"] = AgglomerativeClustering(n_clusters=Agg_num_clusters, 
                                            distance_threshold=Agg_distance_threshold,
                                            affinity='cosine',
                                            linkage='average').fit_predict(dim_PCA)
    ClusterDict["ClusterAgglomerative_tSNE"] = AgglomerativeClustering(n_clusters=Agg_num_clusters, 
                                        distance_threshold=Agg_distance_threshold,
                                        affinity='cosine',
                                        linkage='average').fit_predict(dim_tSNE)
    
    #for the dimension reductions 2,3,4,5 dims:
    num1 = [2,2,3,3,4,4,5,5]
    num2 = [1,2,1,2,1,2,1,2]
    for dim_embedding,nu1,nu2 in zip(dim_embeddings,num1,num2):
        ClusterDict[f"ClusterKmeans_{nu1}_bert{nu2}"] = KMeans(n_clusters=Kmeans_clusters).fit_predict(dim_embedding)
        ClusterDict[f"hdbscan_{nu1}_bert{nu2}"] = hdbscan.HDBSCAN(min_samples = hdbscan_min_samples,
                                              min_cluster_size=hdbscan_min_cluster_size,
                                              cluster_selection_epsilon = 0.5,
                                              metric = 'euclidean',
                                              cluster_selection_method = "eom").fit_predict(dim_embedding)
        
        ClusterDict[f"ClusterAgglomerative_{nu1}_bert{nu2}"] = AgglomerativeClustering(n_clusters=Agg_num_clusters, 
                                                distance_threshold=Agg_distance_threshold,
                                                affinity='cosine',
                                                linkage='average').fit_predict(dim_embedding)
    return ClusterDict
    #, affinity='cosine', linkage='average', distance_threshold=0.4).fit_predict(embeddings)
```

```python
clustering_result = cluster_results(Kmeans_clusters=6,Agg_num_clusters=None,Agg_distance_threshold=1.5)
```

```python
import matplotlib.pyplot as plt
from matplotlib.axes._axes import _log as matplotlib_axes_logger
matplotlib_axes_logger.setLevel('ERROR')
import seaborn as sns
sns.set(style='white', rc={'figure.figsize':(10,8)})


def plot(cluster = clustering_result):    
    
    for clustee_name, clustee in cluster.items():
        clustered = (clustee >= 0)
        if "bert_1" in clustee_name:
            reduced_dim = dim_embedding2_1
        elif "PCA" in clustee_name:
            reduced_dim = dim_PCA
        elif "tSNE" in clustee_name:
            reduced_dim = dim_tSNE
        else : reduced_dim = dim_embedding2_2
            
        print(f"----------{clustee_name}--------------------")

        unique, counts = np.unique(clustee, return_counts=True)
        print(dict(zip(unique, counts)))

        plt.scatter(reduced_dim[~clustered, 0],
                    reduced_dim[~clustered, 1],
                    c=(0.5,0.5,0.5),
                    # s=1,
                    alpha=0.5
                    )
        plt.scatter(reduced_dim[clustered, 0],
                    reduced_dim[clustered, 1],
                    c=clustee[clustered],
                    # s=1,
                    cmap='Spectral')
        plt.show()
```

```python
plot()
```

```python
#inertia

kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(dim_embedding2_2)
                for k in range(1, 10)]
inertias = [model.inertia_ for model in kmeans_per_k]

plt.plot(range(1, 10), inertias, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Inertia", fontsize=14)
plt.show()
```
